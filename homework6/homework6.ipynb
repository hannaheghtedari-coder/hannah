### Prime Star Systems

```python
star_mag = np.array([
[2, 3, 5],
[14, 6, 8],
[11, 13, 17],
[7, 10, 13]
])
print(star_mag)
```

```python
def prime_mag(array):
    def check_prime(n):
        if n < 2:
            return False
        for i in range(2, int(n/2) +1 ):
            if n % i == 0:
                return False # numpy doesn't have a prime function so in theory this tests all numbers from 2 to half of the number as an integer plus one since even if it is closer to the higher integer it rounds down. 
        else: 
            return True

    mask = [any(check_prime(x) for x in row) for row in array] # learned about masked arrays on numpy website to have booleans "interact" with data (so the true/false statements result in a new array)
    return array[mask] 
print(prime_mag(star_mag))
```

### Let's Play Checkers!

```python
checkerboard = np.zeros((8,8))
print(checkerboard)
```

```python
for i in range(checkerboard.shape[0]):
    if i % 2 == 0:
        for j in range(checkerboard.shape[1]):
            if j % 2 == 0:
                checkerboard[i,j] = 1
checkerboard
```

```python
for i in range(checkerboard.shape[0]):
    if i % 2 == 1:
        for j in range(checkerboard.shape[1]):
            if j % 2 == 1:
                checkerboard[i,j] = 1
checkerboard
```

```python
def reverse_checkerboard(array):
    return 1 - array
print(reverse_checkerboard(checkerboard))
```

### The Expanding Universe

```python
universe = np.array(['galaxy', 'clusters'])
def expansion(array, n):
    expanding_space = " " * n
    return np.array([expanding_space.join(string) for string in array])
print(expansion(universe, 4))
```

### Second-Brightest Star

```python
np.random.seed(42) # the answer to the universe because why not
stars = np.random.randint(500, 2000, (5,5)) # random integers from 500 to 2000 using the random seed specified in a 5x5 array
# use np.sort from the previous problem and negative indexing to get second largest for a row of any length! 
stars
```

```python
def second_dimmest(array):
    new_array = []
    for row in array:
        sorted_row = np.sort(row)
        new_array.append(sorted_row[-2]) # adds the second largest value of the row to the new array
    return np.array(new_array)

print(second_dimmest(stars))
```
