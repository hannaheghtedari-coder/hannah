The last line of code is a stylistic thing that technically removes column names (but I still have them when I print, which is good because otherwise they are arbitrary values and unknown meanings), so the shape stays the same.

d) The new shape isn't (11, 20) because after flipping, we don't want a row full of just strings of column names (apart from their actual names, which has to do with indexing and isn't part of the shape). We do want to differentiate by planet, which is the new row index anyway, hence the code that makes it the first column (which would come in handy for manipulations, plotting, etc). The main thing is that each row is filtered by planet and not by variable/ attribute, which again likely will be much easier for when we manipulate the data.
