# HW 7: Plotting with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w*x)
x = np.linspace(0, 2 * np.pi, 1000)
```

```python
amplitude = np.array([0.5, 1, 1.5, 2, 2.5])
frequency = np.array([1, 2, 3, 4, 5])
```

```python
fig, axes = plt.subplots(1, 3, figsize = (25,5))
linestyles = ["-", "--", "-.", ":", (0, (3, 1, 1, 1, 1, 1))]

ls1 = "-" # solid
ls2 = "--" # dashed
ls3 = "-." # dashdot
ls4 = ":" # dotted
ls5 = (0, (3, 1, 1, 1, 1, 1)) # densely dash dotted like __..__..__ etc (info from the matplotlib website)


for A, w in zip(amplitude, frequency):
    if A == 0.5:
        ls = ls1
    elif A == 1:
        ls = ls2
    elif A == 1.5:
        ls = ls3
    elif A == 2:
        ls = ls4
    elif A == 2.5:
        ls = ls5
        
    y1 = make_sine_wave(x, A, 1)
    axes[0].plot(x, y1, label = f"A = {A}", linestyle = ls, color = "dodgerblue")

    y2 = make_sine_wave(x, 1, w)
    axes[1].plot(x, y2, label = f"w = {w}", linestyle = ls, color = "darkturquoise")

    y3 = make_sine_wave(x, A, w)
    axes[2].plot(x, y3, label = f" A = {A} and w = {w}", linestyle = ls, color = "cornflowerblue")


axes[0].set_title("Varying Amplitude", fontsize = 20)
axes[0].set_xlabel("x", fontsize=16)
axes[0].set_ylabel("Asin(x)", fontsize=16)
axes[0].legend(fontsize=14)
axes[1].set_title("Varying Frequency", fontsize = 20)
axes[1].set_xlabel("x", fontsize=16)
axes[1].set_ylabel("sin(wx)", fontsize=16)
axes[1].legend(fontsize=14)
axes[2].set_title("Varying Amplitude and Frequency", fontsize = 20)
axes[2].set_xlabel("x", fontsize=16)
axes[2].set_ylabel("Asin(wx)", fontsize=16)
axes[2].legend(fontsize=14)

plt.tight_layout()
plt.show()
```
