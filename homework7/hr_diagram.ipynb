# Plot a Hertzsprung-Russell Diagram

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Importing Data

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt",
                  skiprows=1)
data
```

### Data Analysis

```python
num_rows, num_cols = data.shape
print(f"There are {num_rows} rows and {num_cols} columns")
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
print(np.min(magnitude), np.max(magnitude))
brightest_star = np.argmin(magnitude)
dimmest_star = np.argmax(magnitude)
```

```python
print(np.min(temperature), np.max(temperature))
hottest_star = np.argmax(temperature)
coolest_star = np.argmin(temperature)
```

```python
print(f"The star with the max mag (dimmest) is in row {dimmest_star + 1}. The star with the min mag (brightest) is in row {brightest_star + 1}.") 
# added +1 because python default indexing starts at zero
```

```python
print(f"The star with the max temp is in row {hottest_star + 1}. The star with the min temp is in row {coolest_star + 1}.") 
# added +1 because python default indexing starts at zero\
```

### Plotting a HR Diagram!

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(log_temperature, log_magnitude, c = log_temperature, cmap = "viridis", label = "Stars")
# default color map is not temperature, so the c = will correct it
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.scatter(log_temperature[brightest_star], log_magnitude[brightest_star],
            color = "black", marker = "*", s = 300, label = "Brightest Star")

plt.scatter(log_temperature[dimmest_star], log_magnitude[dimmest_star],
            color = "purple", marker = "*", s = 300, label = "Dimmest Star")

plt.scatter(log_temperature[hottest_star], log_magnitude[hottest_star],
            color = "green", marker = "*", s = 300, label = "Hottest Star")

plt.scatter(log_temperature[coolest_star], log_magnitude[coolest_star],
            color = "orange", marker = "*", s = 300, label = "Coldest Star")
plt.legend()
plt.xlabel("Temperature (log K)", fontsize=14)
plt.ylabel("Magnitude", fontsize=14)
plt.title("Hertzsprung-Russell Diagram ", fontsize=20)
plt.colorbar(label = "Coldest to Hottest")
plt.show()
```
